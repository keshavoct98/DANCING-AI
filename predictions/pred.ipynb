```python
# Importing required libraries
import numpy as np
import pandas as pd

# Libary required for normalizing data
from sklearn.preprocessing import MinMaxScaler
```

```python
# Importing libraries required for neural networks

from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
```

```python
# Reading Final Dataframe
df = pd.read_csv('../final dataframe/final_df.csv')
df.head()
```

```python
# Removing extra column
df = df.iloc[:,1:]

df.shape
```

```python
# Splitting Training and testing data
# Training data
X = df.iloc[:1449,:36]
y = df.iloc[:1449,36:]

# Testing data
X_test = df.iloc[1449:,:36]
y_test = df.iloc[1449:,36:]

X.shape
```

```python
# Normalizing data for LSTM

sc = MinMaxScaler()
X = sc.fit_transform(X)
X_test = sc.fit_transform(X_test)
y = sc.fit_transform(y)
```

```python
# Reshaping X for LSTM
X = np.reshape(X, (1449,36,1))
```

```python
# Initializing RNN(Recurrent Neural Network)
reg = Sequential()

#Adding network layers        
reg.add(LSTM(units = 72, input_shape = (36,1), return_sequences = True))
reg.add(LSTM(54, activation = 'tanh'))
reg.add(Dense(36))

# Compiling RNN
reg.compile(optimizer = 'adam', loss = 'mean_squared_error')
```

```python
# Fitting RNN to the training set
reg.fit(X, y, batch_size = 16, epochs = 200)
```

```python
# Printing shape of testing dataset
print(X_test.shape)

# Reshaping test set for LSTM
X_test = np.reshape(X_test, (300,36,1))
```

```python
# Predicting test set results
pred = reg.predict(X_test)

pred.shape
```

```python
# Transforming predictions back to real coordinates range
pred_f = sc.inverse_transform(pred)

# Storing predictions in a dataframe
df_f = pd.DataFrame(pred_f)

df_f.head()
```

```python
#Saving dataframe to pred.csv
df_f.to_csv('pred.csv')
```
